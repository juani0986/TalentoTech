```python
import math
print('Ejemplo Calculadora Sencilla (+,-,*,/) por Felipe Escallón:')
'''
ESTO ES UN COMENTARIO MULTILINEA
print('Ejemplo Calculadora Sencilla (+,-,*,/) por Felipe Escallón:\n')
PODEMOS DESCOMENTAR Y PROBAR EL print ANTERIOR
'''
print()
#MENU:
print("MENU:")
print("1: Sumar")
#print("1: +")
print("2: -")
print("3: *")
print("4: Dividir (/)")
print("5: Potenciación")
print("6: Radicación")
print("7: Raiz cuadrada")
print("8: Logaritmo Natural")
print("9: Logaritmo cualquier base")
print("10: Sin")
print("11: Cos")
print("12: Tan")
print()
#Ingreso de datos:
opcion=input("Ingrese la opción deseada del menú: ") # la opción se guarda como cadena
#opcion=int(input("Ingrese la opción deseada del menú: ")) # la opción se guarda como cadena
if opcion == '7':
  numero_uno=float(input('Ingrese el valor para calcular la raiz cuadrada:'))
elif opcion == '10' or opcion == '11' or opcion == '12':
  numero_uno=float(input('Ingrese el valor del angulo:'))
else:
  numero_uno=float(input('Ingrese el primer número de la operación:'))# guarda como flotante (número real)
  numero_dos=float(input('Ingrese el segundo número de la operación:'))# guarda como flotante (número real)
#Posibilidades para ejecutar la calculadora:
#if opcion=='1':
#if opcion=='+':
#if opcion=='Sumar':
if opcion=='1':
  print("Vamos a sumar(+) los dos valores ingresados:")
  print('La suma es:',numero_uno+numero_dos)
if opcion=='2':
  print("Vamos a restar(-) los dos valores ingresados:\n")
  resta=numero_uno-numero_dos
  #print("Resultado="+str(resta)) # voy a concatenar dos cadenas(str1+str2: no suma sino que junta, str1str2)
  print("Resultado=",resta) # voy a concatenar dos cadenas(str1+str2: no suma sino que junta, str1str2)
if opcion=='3':
  print("Vamos a multiplicar(*) los dos valores ingresados:\n")
  #print(f'{numero_uno}*{numero_dos}={numero_uno*numero_dos}')
  mult=numero_uno*numero_dos
  print(f'{numero_uno}*{numero_dos}={mult}')
  #print('{}*{}'.format(numero_uno,numero_dos),"=",'{}'.format(mult))
if opcion=='4'or opcion=='/' or opcion=='Dividir':
  print("Vamos a dividir(/) los dos valores ingresados:\n")
  #VALIDANDO EL DENOMINADOR (debe ser diferente de cero)
  if numero_dos!=0:
    print(numero_uno/numero_dos)
  else:
    print("¡No es posible dividir entre cero!")

if opcion =='5':
  print(f'Calcular la potencia del numero: {numero_uno} elvado a la: {numero_dos} \n')
  potencia=numero_uno**numero_dos
  print("Resultado=",potencia)
if opcion =='6':
  print(f'Calcular la raiz {numero_dos} del número: {numero_uno} \n')
  raiz = numero_uno ** (1/numero_dos)
  print("Resultado=",raiz)   
if opcion =='7':
  print(f'Calcular la raiz cuadrada de: {numero_uno} \n')
  raiz_cuadrada = math.sqrt(numero_uno)
  print("Resultado=",raiz_cuadrada)  
if opcion =='8':
  print(f'Calcular el logaritmo natural de:{numero_uno} \n')
  if numero_uno > 0:
    log_natural = math.log(numero_uno)
    print("Logaritmo natural:", log_natural)
  else:
    print("El número debe ser mayo a 0")  
if opcion == '9':
  print(f'Calcular el logaritmo de:{numero_uno} base {numero_dos} \n')
  if numero_uno > 0:  
    log_base = math.log(numero_uno, numero_dos)
    print("Logaritmo base 5:", log_base)      
  else:
    print("El número debe ser mayo a 0")  
if opcion == '10':    
  print(f'Calcular el seno del angulo:{numero_uno} \n')  
  radianes = math.radians(numero_uno)
  seno = math.sin(radianes)
  grados = math.degrees(seno)
  print("Radianes:", seno)
  print("Grados:", grados)
if opcion == '11':    
  print(f'Calcular el coseno del angulo:{numero_uno} \n')  
  radianes = math.radians(numero_uno)
  coseno = math.cos(radianes)
  grados = math.degrees(coseno)
  print("Radianes:", coseno)
  print("Grados:", grados)
if opcion == '12':    
  print(f'Calcular la tangente del angulo:{numero_uno} \n')  
  radianes = math.radians(numero_uno)
  tangente = math.tan(radianes)
  grados = math.degrees(tangente)
  print("Radianes:", tangente)
  print("Grados:", grados)    
```
