```python
import math

def mostrar_titulo():
  print('Ejemplo Calculadora Sencilla (+,-,*,/) FUNCIONES')

def mostrar_menu():
  print()
  print("MENU:")
  print("1: Sumar (+)")  
  print("2: Restar (-)")
  print("3: Multiplicar (*)")
  print("4: Dividir (/)")
  print("5: Logaritmo (log10)")
  print("6: Potenciación)")
  print("7: Radicación")
  print("8: Coseno (cos)")
  print("9: Seno (sen)")
  print("10: Tangente (tang)")
  print("11: Salir")

def ingresar_datos():
  opcion = int(input("Ingrese la opción: "))
  if opcion in [1, 2, 3, 4, 5, 6, 7]:
      n1 = float(input("Ingrese el primer número: "))
      n2 = float(input("Ingrese el segundo número: "))
  elif opcion in [8, 9, 10]:
      n1 = float(input("Ingrese el número (en grados para funciones trigonométricas): "))
      n2 = 0  
  else:
      n1 = n2 = 0
  return opcion, n1, n2

def calcular_suma(numero_uno,numero_dos):
  suma=numero_uno+numero_dos
  return suma
  
def calcular_resta(numero_uno,numero_dos):
  return numero_uno-numero_dos

def calcular_multiplicacion(numero_uno,numero_dos):
  return numero_uno*numero_dos

def calcular_division(numero_uno,numero_dos):
  return numero_uno/numero_dos

def calcular_logaritmo_base10(numero):
  logaritmo=math.log10(numero)
  return logaritmo

def calcular_coseno(angulo_radianes):
  coseno=math.cos(angulo_radianes)
  return coseno

def calcular_potencia(numero_uno,numero_dos):
  potencia=numero_uno**numero_dos
  return potencia

def calcular_raiz(numero_uno,numero_dos):
  raiz = numero_uno ** (1/numero_dos)
  return raiz

def calcular_seno(angulo_radianes):
  seno=math.sin(angulo_radianes)
  return seno

def calcular_tang(angulo_radianes):
  tang=math.tan(angulo_radianes)
  return tang

def procesar_y_mostrar_operacion(opcion,numero_uno,numero_dos):  
  if opcion==1: 
    print(f"Vamos a sumar(+) los dos valores ingresados: {numero_uno} + {numero_dos}")
    resultado=calcular_suma(numero_uno,numero_dos)
    print('La suma es:',resultado)
  if opcion==2:  
    print(f"Vamos a restar(-) los dos valores ingresados: {numero_uno} + {numero_dos}")
    resultado=calcular_resta(numero_uno,numero_dos)    
    print("Resultado=",resultado) 
  if opcion==3:
    print(f"Vamos a multiplicar(*) los dos valores ingresados: {numero_uno} * {numero_dos}")
    resultado=calcular_multiplicacion(numero_uno,numero_dos)    
    print("Resultado=",resultado)
  if opcion==4:
    print(f"Vamos a dividir(/) los dos valores ingresados:{numero_uno} / {numero_dos}")
    #VALIDANDO EL DENOMINADOR (debe ser diferente de cero)
    if numero_dos!=0:
      resultado=calcular_division(numero_uno,numero_dos)    
      print("Resultado=",resultado)      
    else:
      print("¡No es posible dividir entre cero!")
  if opcion==5: 
    print(f"Vamos a calcular el logaritmo (log10) de los dos valores ingresados:{numero_uno} y {numero_dos}")
    resultado1=calcular_logaritmo_base10(numero_uno)
    resultado2=calcular_logaritmo_base10(numero_dos)    
    print("Resultado1=",resultado1)
    print("Resultado2=",resultado2)

  if opcion==6: 
    print(f'Vamos calcular la potencia del numero: {numero_uno} elvado a la: {numero_dos} \n')
    resultado=calcular_potencia(numero_uno,numero_dos)   
    print("Resultado=",resultado)

  if opcion==7:
    print(f'Calcular la raiz {numero_dos} del número: {numero_uno} \n')
    resultado=calcular_raiz(numero_uno,numero_dos)   
    print("Resultado=",resultado)

  if opcion==8: 
    print("Vamos a calcular el coseno (cos) de los dos valores ingresados:\n")
    resultado1=calcular_coseno(numero_uno)  
    print(f"Coseno de = {numero_uno} Resultado: {resultado1}")

  if opcion==9: 
    print("Vamos a calcular el seno (sen) de los dos valores ingresados:\n")
    resultado1=calcular_seno(numero_uno)
    print(f"Seno de = {numero_uno} Resultado: {resultado1}")

  if opcion==10: 
    print("Vamos a calcular la tangente (tang) de los dos valores ingresados:\n")
    resultado1=calcular_tang(numero_uno)
    print(f"Tangente de = {numero_uno} Resultado: {resultado1}")

def llamar_funciones():
  mostrar_titulo()
  opcion = 0
  while opcion != 11:
    mostrar_menu()
    opcion,numero_uno,numero_dos=ingresar_datos()
    procesar_y_mostrar_operacion(opcion,numero_uno,numero_dos) 

llamar_funciones()

```
