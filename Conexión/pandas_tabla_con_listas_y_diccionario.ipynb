```python
#Import libraries
import pandas as pd #pd es el alias para pandas y se usa en general

#create a dataset - Crear un conjunto de datos
#Let`s build the values lists - costruir la lista de valores
value_n_list = ['Andrea','Felipe','Juan','Andres','Sebastian','Maria','Antonio','Santiago','Valeria','Sofia'] #lista de nombres
value_w_list = [      50,      78,  91.5,      45,         61,   35.5,       85,        15,       45,     38] #[Kg] lista de peso
value_h_list = [    1.76,    1.87,  1.69,     100,       1.85,   1.15,     1.73,      1.05,     1.76,   1.25] #[m] lista de altura

#Let's use a dictionary ({key:value})
#whit the format {"str":list}

data_dict = {"name":value_n_list,"weight":value_w_list,"height":value_h_list}

#Turn a dictionary into a dataframe - convertir un diccionario en un marco de datos
df = pd.DataFrame(data_dict)
#df

df["Edad"] = [    35,    26,  28,     39,       18,   10,     65,      5,     15,   11]

# Agregar la columna 'BMI' (Body Mass Index)
df["BMI"] = df["weight"] / (df["height"] ** 2)

# Insertar la columna en la posición deseada (posición 2)
df.insert(1, "Last name", ["Perez",    "Rodriguez",  "Azuero",     "Garzon",       "Lopez",   "Diaz",     "Caicedo",      "Hurtado",     "Maldonado",   "Morales"] )

df["telefono"] = [11111,3333,444444, 55555,123323,5345345,666666,77777,7877777,8098888]

df
```
