```python
import math

class Calculadora:
    def __init__(self, opcion, num1, num2):
        self.opcion = opcion
        self.num1 = num1
        self.num2 = num2

    @staticmethod
    def mostrar_titulo():
        print('Ejemplo Calculadora Sencilla (+,-,*,/) POO')

    @staticmethod
    def mostrar_menu():
        print("\nMENU:")
        print("1: Sumar (+)")
        print("2: Restar (-)")
        print("3: Multiplicar (*)")
        print("4: Dividir (/)")
        print("5: Salir")

    @staticmethod
    def ingresar_datos():
        opcion = int(input("Ingrese la opción (1-5): "))
        if opcion == 5:
            return opcion, 0, 0  # Salir
        elif 1 <= opcion <= 4:
            n1 = float(input("Ingrese el primer número: "))
            n2 = float(input("Ingrese el segundo número: "))
            return opcion, n1, n2
        else:
            print("Opción no válida")
            return None, None, None

    def calcular_suma(self):
        return self.num1 + self.num2

    def calcular_resta(self):
        return self.num1 - self.num2

    def calcular_multiplicacion(self):
        return self.num1 * self.num2

    def calcular_division(self):
        if self.num2 == 0:
            raise ValueError("No es posible dividir entre cero")
        return self.num1 / self.num2

    def procesar_y_mostrar_operacion(self):
        if self.opcion == 1:
            print(f"\nVamos a sumar(+) los dos valores ingresados: {self.num1} + {self.num2}")
            resultado = self.calcular_suma()
            print('La suma es:', resultado)
        elif self.opcion == 2:
            print(f"\nVamos a restar(-) los dos valores ingresados: {self.num1} - {self.num2}")
            resultado = self.calcular_resta()
            print("Resultado =", resultado)
        elif self.opcion == 3:
            print(f"\nVamos a multiplicar(*) los dos valores ingresados: {self.num1} * {self.num2}")
            resultado = self.calcular_multiplicacion()
            print("Resultado =", resultado)
        elif self.opcion == 4:
            print(f"\nVamos a dividir(/) los dos valores ingresados: {self.num1} / {self.num2}")
            try:
                resultado = self.calcular_division()
                print("Resultado =", resultado)
            except ValueError as e:
                print(e)
        elif self.opcion == 5:
            print("\nSaliendo del programa...")
        else:
            print("\nOpción no válida")

def main():
    Calculadora.mostrar_titulo()
    opcion = 0

    while opcion != 5:
        Calculadora.mostrar_menu()
        opcion, num1, num2 = Calculadora.ingresar_datos()

        if opcion == 5:
            break

        if opcion is not None and num1 is not None and num2 is not None:
            calculadora = Calculadora(opcion, num1, num2)
            calculadora.procesar_y_mostrar_operacion()

if __name__ == "__main__":
    main()
```
