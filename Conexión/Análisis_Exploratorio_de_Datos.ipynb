## Cargar los datos

### Subtask:
Cargar el conjunto de datos "pokemon_data.csv" en un DataFrame de pandas.


**Reasoning**:
Import pandas and load the CSV file into a DataFrame.



```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df_pokemon = pd.read_csv('pokemon_data.csv')

# Display the first few rows to verify the loading
display(df_pokemon.head())
```

## Exploración inicial

### Subtask:
Mostrar las primeras filas, las columnas y sus tipos de datos, y un resumen estadístico del DataFrame.


**Reasoning**:
Display the first few rows, column information, and descriptive statistics of the DataFrame as requested in the instructions.



```python
# Mostrar las primeras 5 filas
display(df_pokemon.head())

# Mostrar información sobre las columnas
display(df_pokemon.info())

# Generar estadísticas descriptivas
display(df_pokemon.describe())
```

## Limpieza de datos

### Subtask:
Verificar si hay valores faltantes o duplicados y manejarlos si es necesario.


**Reasoning**:
Verify null values and duplicated rows, handle nulls in 'moves' column by filling with 'unknown', and drop duplicate rows.



```python
# Check for null values
null_counts = df_pokemon.isnull().sum()
print("Null values per column:")
print(null_counts)

# Check for duplicate rows
duplicate_rows_count = df_pokemon.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_rows_count}")

# Handle null values in the 'moves' column by filling with 'unknown'
df_pokemon['moves'] = df_pokemon['moves'].fillna('unknown')
print("\nNull values after handling 'moves':")
print(df_pokemon.isnull().sum())


# Handle duplicate rows by dropping them
df_pokemon = df_pokemon.drop_duplicates()
print(f"\nNumber of rows after dropping duplicates: {len(df_pokemon)}")
```

## Análisis de la distribución

### Subtask:
Analizar la distribución de algunas variables clave mediante visualizaciones.


**Reasoning**:
Import the necessary libraries for plotting and create histograms for the specified columns to visualize their distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram for base_experience
plt.figure(figsize=(10, 6))
sns.histplot(data=df_pokemon, x='base_experience', kde=True)
plt.title('Distribución de Base Experience')
plt.xlabel('Base Experience')
plt.ylabel('Frecuencia')
plt.show()

# Create a histogram for height
plt.figure(figsize=(10, 6))
sns.histplot(data=df_pokemon, x='height', kde=True)
plt.title('Distribución de Altura')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.show()

# Create a histogram for weight
plt.figure(figsize=(10, 6))
sns.histplot(data=df_pokemon, x='weight', kde=True)
plt.title('Distribución de Peso')
plt.xlabel('Peso')
plt.ylabel('Frecuencia')
plt.show()
```

## Análisis de relaciones

### Subtask:
Explorar la relación entre diferentes variables, como el tipo de Pokémon y sus estadísticas.


**Reasoning**:
Extract the individual statistics from the 'stats' column and create new columns in the DataFrame.



```python
# Extract individual stats from the 'stats' column
stats_df = df_pokemon['stats'].str.split(', ', expand=True)

# Create new columns for each stat
for i in range(stats_df.shape[1]):
    stat_name_value = stats_df[i].str.split('=', expand=True)
    if stat_name_value.shape[1] == 2:
        stat_name = stat_name_value[0].iloc[0]
        df_pokemon[stat_name] = stat_name_value[1].astype(int)

# Display the first few rows with the new columns
display(df_pokemon.head())
```

**Reasoning**:
Create a box plot to visualize the relationship between the first type of Pokémon and their HP, add Spanish titles and labels, and display the plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the first type from the 'types' column
df_pokemon['first_type'] = df_pokemon['types'].str.split(', ', expand=True)[0]

# Create a box plot of HP by the first type
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_pokemon, x='first_type', y='hp')

# Add titles and labels in Spanish
plt.title('Relación entre el Primer Tipo de Pokémon y sus Puntos de Salud (HP)')
plt.xlabel('Primer Tipo de Pokémon')
plt.ylabel('Puntos de Salud (HP)')
plt.xticks(rotation=90) # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```
