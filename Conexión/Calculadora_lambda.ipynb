```python
import math

def mostrar_titulo():
  print('Ejemplo Calculadora Sencilla (+,-,*,/) FUNCIONES')

def mostrar_menu():
  print()
  print("MENU:")
  print("1: Sumar (+)")  
  print("2: Restar (-)")
  print("3: Multiplicar (*)")
  print("4: Dividir (/)")
  print("5: Logaritmo (log10)")
  print("6: Potenciación)")
  print("7: Radicación")
  print("8: Coseno (cos)")
  print("9: Seno (sen)")
  print("10: Tangente (tang)")
  print("11: Salir")

def ingresar_datos():
  opcion = int(input("Ingrese la opción: "))
  if opcion in [1, 2, 3, 4, 5, 6, 7]:
      n1 = float(input("Ingrese el primer número: "))
      n2 = float(input("Ingrese el segundo número: "))
  elif opcion in [8, 9, 10]:
      n1 = float(input("Ingrese el número (en grados para funciones trigonométricas): "))
      n2 = 0  
  else:
      n1 = n2 = 0
  return opcion, n1, n2

def calcular_suma(numero_uno,numero_dos):
  suma=numero_uno+numero_dos
  return suma
  
def calcular_resta(numero_uno,numero_dos):
  return numero_uno-numero_dos

def calcular_multiplicacion(numero_uno,numero_dos):
  return numero_uno*numero_dos

def calcular_division(numero_uno,numero_dos):
  return numero_uno/numero_dos

def calcular_logaritmo_base10(numero):
  logaritmo=math.log10(numero)
  return logaritmo

def calcular_coseno(angulo_radianes):
  coseno=math.cos(angulo_radianes)
  return coseno

def calcular_potencia(numero_uno,numero_dos):
  potencia=numero_uno**numero_dos
  return potencia

def calcular_raiz(numero_uno,numero_dos):
  raiz = numero_uno ** (1/numero_dos)
  return raiz

def calcular_seno(angulo_radianes):
  seno=math.sin(angulo_radianes)
  return seno

def calcular_tang(angulo_radianes):
  tang=math.tan(angulo_radianes)
  return tang

def procesar_y_mostrar_operacion(opcion,numero_uno,numero_dos):  
  if opcion==1: 
    suma = lambda a,b: a+b
    resultado = suma(numero_uno, numero_dos)
    print(f"La suma es: {resultado}")

  if opcion==2:  
    resta = lambda a,b: a-b
    resultado = resta(numero_uno, numero_dos)
    print(f"La resta es: {resultado}")

  if opcion==3:
    multiplicacion = lambda a,b: a*b
    resultado = multiplicacion(numero_uno, numero_dos)
    print(f"La multiplicación es: {resultado}")

  if opcion==4:  
    division = lambda a,b: a/b
    resultado = division(numero_uno, numero_dos) if numero_dos!=0 else "No se puede dividir entre 0"
    print(f"La division de {numero_uno} entre {numero_dos} es: {resultado}")

  if opcion==5: 
    logaritmo = lambda a,b: math.log(a, b)
    resultado = logaritmo(numero_uno, numero_dos)
    print(f"El logaritmo de {numero_uno} base {numero_dos} es igual: {resultado}")

  if opcion==6: 
    potencia = lambda base,exponente: math.pow(base,exponente)# base**exponente
    resultado = potencia(numero_uno, numero_dos)
    print(f"La potencia de base {numero_uno} y exponente {numero_dos} es igual: {resultado}")

  if opcion==7:
    raiz = lambda a,b: a ** (1/b)
    resultado = raiz(numero_uno, numero_dos)
    print(f"La raiz de {numero_uno} a la {numero_dos} es igual: {resultado}")

  if opcion==8: 
    coseno = lambda angulo_radianes: math.cos(angulo_radianes)
    resultado = coseno(numero_uno)
    print(f"El coseno del angulo_radianes: {numero_uno} es igual: {resultado}")

  if opcion==9: 
    seno = lambda angulo_radianes: math.sin(angulo_radianes)
    resultado = seno(numero_uno)
    print(f"El seno del angulo_radianes: {numero_uno} es igual: {resultado}")

  if opcion==10: 
    tangente = lambda angulo_radianes: math.tan(angulo_radianes)
    resultado = tangente(numero_uno)
    print(f"La tangente del angulo_radianes: {numero_uno} es igual: {resultado}")

def llamar_funciones():
  mostrar_titulo()
  opcion = 0
  while opcion != 11:
    mostrar_menu()
    opcion,numero_uno,numero_dos=ingresar_datos()
    procesar_y_mostrar_operacion(opcion,numero_uno,numero_dos) 

llamar_funciones()
```
